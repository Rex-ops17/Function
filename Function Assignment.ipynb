```python
# PRACTICAL QUESTION

# Answer 1
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)
# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(numbers))  # Output: 12 (2 + 4 + 6)


# Answer 2
def reverse_string(s):
    return s[::-1]
# Example usage:
print(reverse_string("hello"))  # Output: "olleh"


# Answer 3
def square_numbers(numbers):
    return [num ** 2 for num in numbers]
# Example usage:
print(square_numbers([1, 2, 3, 4]))  # Output: [1, 4, 9, 16]


# Answer 4
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")  # Output: List of prime numbers between 1 and 200


#Answer 5
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count < self.terms:
            value = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return value
        else:
            raise StopIteration
# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")  # Output: Fibonacci sequence up to 10 terms


#Answer 6
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
# Example usage:
for power in powers_of_two(5):
    print(power, end=" ")  # Output: 1 2 4 8 16 32


#Answer 7
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()  # .strip() to remove newline characters
# Example usage:
# Make sure to create a test file "example.txt" before running this
# for line in read_file_line_by_line("example.txt"):
#     print(line)


#Answer 8
my_list = [(1, 2), (3, 1), (5, 4)]
sorted_list = sorted(my_list, key=lambda x: x[1])
# Example usage:
print(sorted_list)  # Output: [(3, 1), (1, 2), (5, 4)]


#Answer 9
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
# Example usage:
celsius_temps = [0, 10, 20, 30]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)  # Output: [32.0, 50.0, 68.0, 86.0]


#Answer 10
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda x: x not in vowels, s))
# Example usage:
input_string = "Hello World"
print(remove_vowels(input_string))  # Output: "Hll Wrld"


#Answer 11
# Input list of orders
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einf√ºhrung in Python3, Bernd Klein", 3, 24.99],
    ]


# Using map and lambda
result = list(map(
    lambda order: (order[0], round(order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0), 2)),
    orders
))

# Output the result
print(result)

# Result
[(34587, 163.8), (98762, 284.0), (77226, 108.85), (88112, 84.97)]
















```
